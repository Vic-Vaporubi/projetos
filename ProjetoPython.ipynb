Seno, cosseno e tangente do ângulo:

```python
#Funções seno, cosseno e tangente.

#Importação de funções da biblioteca:

from math import sin, cos, tan, pi

#definição das funções:

def sinG(ang):
  ang = ang * pi / 180
  ang = sin(ang)
  return ang

def cosG(ang):
  ang = ang * pi / 180
  ang = cos(ang)
  return ang

def tanG(ang):
  ang = ang * pi / 180
  ang = tan(ang)
  return ang

#ang = 30

print(f"Com um ângulo de 30°, temos, respectivamente, seno, cosseno e tangente,\n {sinG(30)},\n {cosG(30)},\n {tanG(30)}.")
```
